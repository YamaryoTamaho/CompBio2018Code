# 対数らせん
$$
r(\theta)=r_0 e^{a \theta}
$$
に基づく様々な巻きパタンを記述する．

## 準備
パッケージの読み込み

```python
#　パッケージの読み込み
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 対数らせんプロット用関数
def plotLS(x, y):
    fig = plt.figure(dpi=200)
    ax = fig.gca()
    ax.set_aspect('equal')
    ax.plot(x,y)  
    plt.show()

# 対数らせんファイル出力用関数
def savefigLS(x, y, filename):
    fig = plt.figure(dpi=200)
    ax = fig.gca()
    ax.set_aspect('equal')
    ax.plot(x,y)
    plt.savefig(filename)
```

## 計算とプロット

```python
thetaRange = np.linspace(0, 16*np.pi, 3600)
a = 0.05
r0 = 1

x = []
y = []
for theta in thetaRange:
    r = r0*np.exp(a*theta)
    x.append(r*np.cos(theta))
    y.append(r*np.sin(theta))
```

```python
# プロット
plotLS(x,y)
```

```python
savefigLS(x,y,"a=0.05.png")
```
